```python
import os
from dotenv import load_dotenv
from google import genai

load_dotenv()

client=genai.Client()
model_name = "gemini-2.5-flash"
```

```python
from google.genai import types

response = client.models.generate_content(
    model=model_name,
    contents=[
        types.Content(
            role="user",
            parts=[types.Part.from_text(text="Which is the best engineering college in Asia? Answer in one sentence.")]
        )
    ]
)
```
